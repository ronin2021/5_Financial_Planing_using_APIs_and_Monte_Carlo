## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Add format specifier to request url
btc_url = btc_url + "?format=json"
eth_url = eth_url + "?format=json"

# Execute get request
requests.get(btc_url)
requests.get(btc_url)

# Execute GET request and store response
btc_response_data = requests.get(btc_url)
eth_response_data = requests.get(eth_url)
```

```python
# Formatting as json
btc_data = btc_response_data.json()
eth_data = eth_response_data.json()

# Add indents to JSON and output to screen
print(json.dumps(btc_data, indent=4))
print(json.dumps(eth_data, indent=4))
```

```python
# Fetch current BTC price
btc_current_value = btc_data["data"]["1"]["quotes"]["USD"]["price"]
print("Current value of a unit of BTC: $" + str(btc_current_value))

# Fetch current ETH price
eth_current_value = eth_data["data"]["1027"]["quotes"]["USD"]["price"]
print("Current value of a unit of ETH: $" + str(eth_current_value))

# Compute current value of my crpto
my_btc_value = my_btc * btc_current_value
my_eth_value = my_eth * eth_current_value

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Current amount of shares
my_spy = 50
my_agg = 200
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")


# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
# Format current date as ISO format
# Set today date
today_date = pd.Timestamp("2021-06-22", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# Get 1 year's worth of historical data for AGG and SPY
df_ticker = alpaca.get_barset(
    tickers,
    timeframe,
    start=today_date,
    end=today_date,
    limit=1000,
).df

# Preview DataFrame
df_ticker.head()
```

```python
# Pick AGG and SPY close prices
agg_close_price = df_ticker["AGG"]["close"][0]
spy_close_price = df_ticker["SPY"]["close"][0]

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_agg_value = my_agg * agg_close_price
my_spy_value = my_spy * spy_close_price

# Print current value of share
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthley_income = 12000

# Total value of crypto
my_total_crypto_value = my_btc_value + my_eth_value


# Total value of shares
my_total_share_value = my_agg_value + my_spy_value

# Create savings DataFrame
savings_dict = { 
                "amount" : { "crypto" : my_total_crypto_value, "shares" : my_total_share_value }
                }

savings_df = pd.DataFrame.from_dict(savings_dict)
# Display savings DataFrame
display(savings_df.head())
```

```python
# Plot savings pie chart
savings_df.plot.pie(y = "amount", figsize = (5 , 5), autopct='%1.1f%%')
plt.show()
```

```python
# Set ideal emergency fund
emergency_fund = monthley_income * 3

# Calculate total amount of savings
total_savings = my_total_crypto_value + my_total_share_value

# Validate saving health
if total_savings > 3 * monthley_income:
    print("congratulations, you have enough money in your emergency fund!")
elif total_savings == 3 * monthley_income:
    print("congratulations, you have reached your financial goal of an adequate emergency fund!")
else:
    print(f"you are ${round(3* monthley_income - total_savings, 2): ,} away from having a necessary emmergency fund!")     
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-06-18', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-06-18', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# Get 1 year's worth of historical data for AGG and SPY
stock_data_df = alpaca.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df

# Preview DataFrame
stock_data_df.head(3)
```

```python
# Set index to reflect only the date
stock_data_df.index = stock_data_df.index.date
stock_data_df.dropna(inplace = True)

stock_data_df.head(3)
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# Set number of simulations
num_sims = 500

# Configure a Monte Carlo simulation to forecast thirty years daily returns
MC_thirty_years = MCSimulation(
    portfolio_data = stock_data_df,
    weights = [.4, .6],
    num_simulation = num_sims,
    num_trading_days = 252*30
)

# Printing the simulation input data
MC_thirty_years.portfolio_data.head(3)
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_thirty_years.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_thirty_years.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_thirty_years.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
thrty_yr_tbl = MC_thirty_years.summarize_cumulative_return()

# Print summary statistics
print(thrty_yr_tbl)
```

### Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
thrty_yr_ci_lower = round(thrty_yr_tbl[8] * initial_investment, 2)
thrty_yr_ci_upper = round(thrty_yr_tbl[9] * initial_investment, 2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment: ,} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${thrty_yr_ci_lower: ,} and ${thrty_yr_ci_upper: ,}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
thrty_yr_50pcent_more_principle_ci_lower = round(thrty_yr_tbl[8] * initial_investment, 2)
thrty_yr_50pcent_more_principle_ci_upper = round(thrty_yr_tbl[9] * initial_investment, 2)
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment: ,} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${thrty_yr_50pcent_more_principle_ci_lower: ,} and ${thrty_yr_50pcent_more_principle_ci_upper: ,}")
```

## Optional Challenge - Early Retirement


### Five Years Retirement Option

```python
# Configuring a Monte Carlo simulation to forecast 5 years cumulative returns
# Set number of simulations
num_sims = 1000

# Configure a Monte Carlo simulation to forecast five years daily returns
MC_five_yrs = MCSimulation(
    portfolio_data = stock_data_df,
    weights = [.1, .9], # Increase the stocks to 80% of the portflio
    num_simulation = num_sims,
    num_trading_days = 252*5 # 5-year simulation
)

# Printing the simulation input data
MC_five_yrs.portfolio_data.head(3)
```

```python
# Running a Monte Carlo simulation to forecast 5 years cumulative returns
MC_five_yrs.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_five_yrs.plot_simulation()
```

### Plot the Simulated Daily Returns Trajectory for the Portfolio over the Next Five Years (252 * 5 = 1260 Trading Days)

```python
# Compute summary statistics from the simulated daily returns of the portfolio
five_yrs_simulated_returns_data = {
    "mean": list(MC_five_yrs.simulated_return.mean(axis=1)),
    "median": list(MC_five_yrs.simulated_return.median(axis=1)),
    "min": list(MC_five_yrs.simulated_return.min(axis=1)),
    "max": list(MC_five_yrs.simulated_return.max(axis=1))
}

# Create a DataFrame with the summary statistics
simulated_rtrns_five_yrs_df = pd.DataFrame(five_yrs_simulated_returns_data)

# Display sample data
simulated_rtrns_five_yrs_df.head(3)
```

```python
# Use the `plot` function to visually analyze the trajectory of portfolio daily returns on the next five years of trading days simulation
simulated_rtrns_five_yrs_df.plot(title="Simulated Daily Returns Behavior of Portfolio Over the Next Five Years\n", ylabel = "Cumulative Returns", xlabel = "Number of Trading Days")
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_five_yrs.plot_distribution()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
five_yrs_tbl = MC_five_yrs.summarize_cumulative_return()

# Print summary statistics
print(five_yrs_tbl)
```

```python
# Set initial investment
initial_investment = 80000 # increased principle to $80,000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $60,000
five_yrs_ci_lower = round((five_yrs_tbl[8] * initial_investment), 2)
five_yrs_ci_upper = round((five_yrs_tbl[9] * initial_investment), 2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment: ,} in the portfolio"
      f" over the next 5 years will end within in the range of"
      f" ${five_yrs_ci_lower: ,} and ${five_yrs_ci_upper: ,}")
```

### Calculate the Simulated Profits/Losses of $80,000 Investment in the portfolio Over the Next Ten Years

```python
# Multiply an initial investment (= $80,000) by the daily returns of simulative stock prices to return the progression of daily returns in terms of money

cumulative_five_yrs_pnl = initial_investment * simulated_rtrns_five_yrs_df

# Display sample data
cumulative_five_yrs_pnl.head(3)
```

```python
# Use the `plot` function to visually analyze the trajectory of portfolio daily returns on the next ten years of trading days simulation
cumulative_five_yrs_pnl.plot(title="Simulated Cumulative Profit & Loss of $80,000 Principle Invested in the Portfolio Over the Next five Years\n", xlabel = "Number of Days", ylabel = "Cumulative Returns in US$")
```
